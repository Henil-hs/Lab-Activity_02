```python
#building Ml model for breast cancer data
# importing the libraries
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv("/content/breast-cancer-wisconsin.csv", header = None)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
x = df.iloc[:,:10].values
y = df.iloc[:,10:].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20) 
```

```python
#Create the models
#Create the 1inear regression model
lg = LogisticRegression().fit(x_train, y_train)
```

```python
#Show the model linear regression prediction
lg_prediction = lg.predict(x_test)
print(lg_prediction)
```

```python
mse = mean_squared_error(lg_prediction,y_test)

rmse = math.sqrt(mse)

print(rmse)
```
